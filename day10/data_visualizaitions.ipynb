```python
# Line Plot:
# Write a Python program to create a simple line plot of a mathematical function (e.g., sine wave) using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
countries = ['USA', 'Canada', 'Germany', 'UK', 'France']
population = [331, 38, 83, 67, 65]  # in millions

plt.bar(countries, population)
plt.title('Population of Countries')
plt.xlabel('Country')
plt.ylabel('Population (millions)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
df = sns.load_dataset('iris')

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
df = sns.load_dataset('iris')

sns.histplot(df['sepal_length'], bins=10, kde=True)
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Display the first few rows of the dataset to ensure it loaded correctly
print(df.head())

# Calculate the correlation matrix, excluding the 'species' column
corr = df.drop(columns='species').corr()

# Display the correlation matrix to ensure it was calculated correctly
print(corr)

# Create the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Iris Dataset Correlations')
plt.show()
```
